### Duplicate data

The next cell is responsible for dividing the **duplicates rows** from the unique ones. To do that the function *find_duplicates_android* loops through the valid rows of the Android dataset and appends to the unique apps list only rows whose **"name" section** never appeared before, otherwise it appends the row to the duplicate list.

To have a more intelligent algorithm and don't remove randomly the duplicates, we'll only keep the row with the **highest number of reviews** since the higher the number of reviews, the **more recent** the data should be. 

```python
duplicate_apps_android = []
unique_apps_android = []

def find_duplicates_android(android):
    for row in android[1:]:
        name = row.split(',')[0]
        reviews=row.split(',')[3]
        print(name)
        print(reviews+'\n')
        if name in unique_apps_android and reviews > row.split(',')[3]:
            duplicate_apps_android.append(name)
            
        unique_apps_android.append(name)
        
find_duplicates_android(android_apps_data) 
print(len(duplicate_apps_android))
print(len(unique_apps_android))
print(duplicate_apps_android[:20])
```
